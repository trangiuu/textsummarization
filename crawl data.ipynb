```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

Chỉnh cái getinfor này, thêm output chủ đề ý

```python
#hàm lấy thông tin: tiêu đề, nội dung, tag, nội dung thu gọn (sapo)
def getinfor(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    #tiêu đề
    tieude = soup.find("h1")
    tieude_final = tieude.text.strip() if tieude else "No title found"
    #sapo
    p_des=soup.find("p",class_="description")
    sapo = ''.join([text for text in p_des if not text.name == 'span']) if p_des else "No description found"

    # if p_des:
    #         sapo = ''.join([text for text in p_des.stripped_strings if not isinstance(text, str)])
    # else:
    #         sapo = "No description found"

    #nội dung
    noidung=soup.find_all("p",class_="Normal")
    e=soup.find_all("strong")
    for strong_tag in e:
        for p_tag in noidung:
            if strong_tag in p_tag:
                strong_tag.decompose()

    noidung_khongtacgia = ' '.join([p.get_text() for p in noidung])
    row=[]
    row.append(tieude_final)
    row.append(sapo)
    row.append(noidung_khongtacgia)
    list_row=[row[0],row[1],row[2]]
    return list_row
```

```python
#từ trang chủ sang các trang chủ đề
#ví dụ từ trang chủ-> thời sự/ thể thao,...
def geturl01(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    nav_list=[]
    nav_list=soup.find_all("nav",class_="main-nav")

    li_list=[]
    for i in range (len(nav_list)):
        for x in nav_list[i].find_all("li"):
            li_list.append(x)


    links_01=[]
    for i in range (len(li_list)):
        for x in li_list[i].findAll("a"):
            href=x.get("href")
            if("https:" not in href and len(href)!=1 and "java" not in href and "tin-tuc" not in href):
                href="https://vnexpress.net"+href
                links_01.append(href)
    return links_01
```

```python
links01_demo=geturl01('https://vnexpress.net')
print(links01_demo)
```

```python
#từ chủ đề sang phân loại chủ đề
# từ thời sự-> thời sự/dân sinh hoặc thời sự/chính trị
def geturl02(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    li_link=[]
    ul_links = soup.find_all("nav")
    for i in range (len(ul_links)):
        for x in ul_links[i].find_all("ul",class_="ul-nav-folder"):
            li_link.append(x)


    links_02=[]
    for i in range (len(li_link)):
        for x in li_link[i].find_all("a"):
            href=x.get("href")
            if("https://vnexpress.net" in href):
                links_02.append(href)
            else:
                href='https://vnexpress.net/'+href
                links_02.append(href)
    return links_02
    
```

```python
link02_demo=geturl02('https://vnexpress.net/thoi-su')
print(link02_demo)
```

```python
#đang cấn chỗ này, vì không tìm thấy link bài báo trong link 
#mục tiêu là tìm được link href có đuôi html ấy, nhưng soupfind thì ko có ra
#fix được chỗ này thì ngon
#đầu vào là output của geturl02
#output là link báo, đuôi html. Output hàm url03 là input hàm getinfor

def geturl03(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    links = soup.find_all('a')
    titles = set()  

    for link in links:
        href = link.get("href")
        if href and "html" in href and "#" not in href:
            titles.add(href)

    return list(titles)  

```

Chỗ này là do không lấy được link từ hàm geturl03 nên là tôi lấy luôn link trong txt url của ông trên mạng

```python
import os

def lay_ten_file_txt(folder_path):
    files_list = []
    for file in os.listdir(folder_path):
        if file.endswith(".txt"):
            files_list.append(file)
    return files_list

folder_path = "url"
ten_file_txt = lay_ten_file_txt(folder_path)
print(ten_file_txt)
```

```python
def gethtml(ten):
    with open(ten,'r') as file:
        lines = file.readlines()
    lines = [line.strip() for line in lines]
    return list(lines)
    
```

```python
html=[]
for i in range(len(ten_file_txt)):
    html_ele=gethtml(ten_file_txt[i])
    for x in range(len(html_ele)):
        html.append(html_ele[x])
```
